## Introduction
- Ford GoBike is the firts regional and large scale bicycle sharing system covering the greater San Francisco Bay area. This syste was established in the year 2013. As of 2018 they had about 2600 bicycles in 262 sations across San Francisco, East Bay and San Jose.

- The bicycles are in two categories; the classic dock model and the hybrid model that can be parked at a dock or used in a dockless manner. They are usually available 24 Hrs a day, 365 days a year. Customers may choose either a single ride or an annual membership. Single rides start at around 3 dollars per single trip and a day may cost up to 10 dollars. Memberships cost 25 dollars per month or 159 dollars per year.

## 3. Cleaning the data

### Define
#### Changing the data types
 - bike_id, start_station_id, end_station_id to object format.
 - start and end times to timestamp format.
 - user_type,gender,bike_share_for_all_trip to category format

### Code

```python
# create a copy of the dataset
df_clean = df.copy()
# changing the format to datetime format
df_clean.start_time = pd.to_datetime(df_clean.start_time) 
df_clean.end_time = pd.to_datetime(df_clean.end_time) 
# changing bike_id, start_station_id, end_station_id to object format.
df_clean.bike_id = df_clean.bike_id.astype(str)
df_clean.start_station_id = df_clean.start_station_id.astype(str)
df_clean.end_station_id = df_clean.end_station_id.astype(str)
# changing user_type,gender,bike_share_for_all_trip to category format
df_clean.user_type = df_clean.user_type.astype('category')
df_clean.member_gender = df_clean.member_gender.astype('category')
df_clean.bike_share_for_all_trip = df_clean.bike_share_for_all_trip.astype('category')
```

### Test

```python
df_clean.info()
```

### Define
**Calculating the age of the members**

### Code

```python
# calculating the age of the members by subtracting their birth year from the current year 2019
df_clean['member_age'] = 2019 - df_clean['member_birth_year']
```

### Test


```python
df_clean.head(20)
```

```python
# getting the statistics of the users age
df_clean.member_age.describe()
```

**From our results we can see that 75% of the users are under the age 39**

```python
# further checking where the bigger percentage lies
df_clean.member_age.describe(percentiles=[.99])
```

**We can see that 99% of the user are of 64 and below.**

```python
# we get to keep records of users with the age 64 and below
df_clean = df_clean.query('member_age <= 64')
```

```python
# changing the datatype of the members age and year
df_clean.member_birth_year = df_clean.member_birth_year.astype(int)
df_clean.member_age = df_clean.member_age.astype(int)
```

```python
# testing our code
df_clean.info()
```

### Define
**Further adding the Hrs and Days in our data**

```python
# extracting the hours
df_clean['start_time_hour'] = df_clean['start_time'].dt.hour
df_clean['end_time_hour'] = df_clean['end_time'].dt.hour
```

```python
# extracting the minutes from the seconds
df_clean['duration_min'] = df_clean['duration_sec']/60
#changing the datatype
df_clean.duration_min = df_clean.duration_min.astype(int)
```

```python
# extracting the weekday
df_clean['start_day_of_week']=df_clean['start_time'].dt.strftime('%a')
```
